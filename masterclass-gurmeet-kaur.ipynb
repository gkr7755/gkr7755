```python
#boxplot
import seaborn as sbn
```

```python
sbn.boxplot('MonthlyCharge', data=TC)
```

```python
q1=TC['MonthlyCharge'].quantile(0.25)
q3=TC['MonthlyCharge'].quantile(0.75) 
```

```python
iqr= q3-q1
```

```python
iqr
```

```python
tc=TC[(TC['MonthlyCharge']>=q1-1.5*iqr) & (TC['MonthlyCharge']<=q3+1.5*iqr)]
```

```python
tc
```

```python
sbn.boxplot('MonthlyCharge', data=tc)
```
